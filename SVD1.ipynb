# Singular Value Decomposition

## Zadanie 2

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc

def reduce(k, U, s, Vh):
    return U[:, :k], s[:k], Vh[:k, :]

def combine(U, s, Vh):
    D = np.diag(s)
    return (U.dot(D)).dot(Vh)

def compress(x, k):
    U, s, Vh = np.linalg.svd(x)
    U_red, s_red, V_red = reduce(k, U, s, Vh)
    return combine(U_red, s_red, V_red)
```

```python
def test(i):
    ascent = scipy.misc.ascent()
    compressed = compress(ascent, i)
    plt.imshow(compressed)
    plt.show()
```

```python
for i in range(10, 100, 10):
    test(i)
```
