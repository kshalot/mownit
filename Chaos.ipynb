Uk≈Çad Lorenza:

$$\begin{split}\begin{aligned}
\dot{x} & = \sigma(y-x) \\
\dot{y} & = \rho x - y - xz \\
\dot{z} & = -\beta z + xy
\end{aligned}\end{split}$$

gdzie:<br>
ùúé - liczba Prandtla,<br>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D


def lorenz_system(V, t, sigma, beta, rho):
    x, y, z = V
    return (sigma * (y - x),
           rho * x - y - x * z,
           -beta * z + x * y)


sigma = 10
beta = 8/3
rho = 28.5
n = 10000

ls = np.linspace(0, 100, n)
eq = odeint(lorenz_system, (1, 2, 3), ls, args=(sigma, beta, rho))
x, y, z = eq.T

fig = plt.figure()
ax = fig.gca(projection='3d')

ls = np.linspace(0, 1, n)
gap = 100
for i in range(0, n, 10):
    ax.plot(x[i:i + gap + 1], y[i:i + gap + 1], z[i:i + gap + 1], color=(0, ls[i], 1))

plt.show()
```
